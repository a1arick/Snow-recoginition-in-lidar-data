```python
import torch
import sys
import pandas as pd
import numpy as np
from tqdm import tqdm

sys.path.append('/Users/alexander.savelyev/PycharmProjects/Snow_recognition')
from PointNet import PointNetDenseCls
from PointNet import PointNetDenseCls
```
